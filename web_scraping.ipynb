```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import numpy as np
```

```python
url='https://www.rokomari.com/book/ebook?ref=sm_p3'
```

```python
web_code=requests.get(url).text
```

```python
soup=BeautifulSoup(web_code,'lxml')
```

```python
soup
```

```python
soup.prettify()
```

```python
print(soup.prettify())
```

```python
code_data=soup.find_all('p',class_='book-title')
```

```python
code_data
```

```python
l=len(soup.find_all('p',class_='book-title'))
```

```python
print(l)
```

```python
book_name=[]
for i in code_data:
  book_name.append(i.get_text(strip=True))
```

```python
book_name
```

```python
code_writer_data=soup.find_all('p',class_='book-author')
```

```python
writer_name=[]

for i in code_writer_data:
  writer_name.append(i.get_text(strip=True))
```

```python
writer_name
```

```python
code_original_price_data=soup.find_all('strike',class_='original-price')
```

```python
original_price=[]

for i in code_original_price_data:
  original_price.append(i.text)
```

```python
original_price
```

```python
code_price=soup.find_all('span')
```

```python
price=[]

for i in code_price:
  price.append(i.text)
```
