```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/dataset/Sleep_health_and_lifestyle_dataset.csv')
```

```python
df
```

```python
df["Sleep Health Risk Index"] = (df["Stress Level"] + (10 - df["Quality of Sleep"]))/2
```

```python
df
```

```python
df["Sleep Health Risk Index"] = df['Sleep Health Risk Index'].clip(0,10)
```

```python
df
```

```python
def categorize_risk(shri):
  if shri<3:
    return "Low Risk"
  elif 3<=shri<6:
    return "Moderate Risk"
  else:
    return "High Risk"

df["Sleep Health Risk Category"] = df["Sleep Health Risk Index"].apply(categorize_risk)
```

```python
df
```

# add another feature in the index

```python
bmi_mapping = {
    "Normal" : 1,
    "Normal Weight" : 1,
    "Overweight" : 2,
    "Obese" : 3
}

df["BMI Value"] = df["BMI Category"].map(bmi_mapping)
```

```python
df
```

# index 2

```python
df["Sleep Health Risk Index 2"] = ((df["Stress Level"] + (10 - df["Quality of Sleep"]))/2) * df["BMI Value"]
```

```python
df
```

```python
df["Sleep Health Risk Index 2"] = df["Sleep Health Risk Index 2"].clip(0,10)
```

```python
df
```

```python
def categorize_risk(shri):
  if shri<5:
    return "Low Risk"
  elif 5<=shri<=8:
    return "Moderate Risk"
  else:
    return "High Risk"

df["Sleep Health Risk Category 2"] = df["Sleep Health Risk Index 2"].apply(categorize_risk)
```
